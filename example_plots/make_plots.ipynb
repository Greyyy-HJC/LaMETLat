### Fig. 6

Here only the plot of this work is shown, results from other groups can be found in references.

```python
fig6 = gv.load("cache/fig6.dat")

fig, ax = default_plot()

#! This work
ax.errorbar(fig6["b_fm"], fig6["csk_mean"], yerr=fig6["csk_sdev"], label="This Work", fmt=marker_ls[0], color="red", **errorb)

ax.set_xlabel(r"$b_\perp$ [fm]", **fs_p)
ax.set_xlim([0, 0.8])
ax.set_ylabel(r"$\gamma^{\overline{\mathrm{MS}}}(b_\perp;~\mu)$", **fs_p)
ax.set_ylim([-1.6, 1.2])
plt.legend(ncol=4, fontsize=11)
plt.tight_layout()
plt.savefig("plots/fig6.pdf", transparent=True)
plt.show()
```

### Fig. 8

```python
fig8 = gv.load("cache/fig8.dat")

mom_ls = [8, 9, 10]
mom_unit = 0.43
ff_pz_gev = 2.58

fig, ax = default_plot()

for mom in mom_ls:
    pz = mom * mom_unit
    label = r"$P_{\mathrm{WF}}=$" + f"{pz:.2f} GeV, " + r"$P_{\mathrm{FF}}=$" + f"{ff_pz_gev} GeV"
    
    ax.errorbar(np.array(fig8[f"PWF={pz:.2f}_PFF={ff_pz_gev}_bfm"])+0.01, fig8[f"PWF={pz:.2f}_PFF={ff_pz_gev}_mean"], yerr=fig8[f"PWF={pz:.2f}_PFF={ff_pz_gev}_sdev"], label=label, **errorb)

ax.errorbar(fig8["final_bfm"], fig8["final_mean"], fig8["final_sdev"], label='Final results', marker='x', elinewidth=2, markersize=5, capsize=3, linestyle='none')

ax.plot(fig8["1loop_bfm"], fig8["1loop_mean"], label='1-loop Fix', color='red', linestyle='--')
ax.plot(fig8["1loop_rgr_bfm"], fig8["1loop_rgr_mean"], label='1-loop RGR LL', color='blue', linestyle='--')
ax.set_xlabel(r"$b_\perp$ [fm]", **fs_p)
ax.set_ylabel(r"$S^{\mathrm{ratio}}_I(b_\perp;~\mu)$", **fs_p)
ax.set_ylim([-0.1, 1.8])
plt.legend(ncol=2, fontsize=11)
plt.tight_layout()
plt.savefig("plots/fig8.pdf", transparent=True)
plt.show()
```
